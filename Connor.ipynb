```python
import matplotlib.pyplot as plt
import numpy as np
import random

class World: 
    def __init__(self, rows=50, cols=50, environment='placeholder', animals=[]):
        """
        Initializes a Board.

        Args:
            rows/cols: The size of the world
            
            environment: Choose from (Jungle, Forest, Meadow, Savannah, Desert)
                - Jungle has most vegetation/water
                - Desert has least
                
            animals: A list of all of the animals in the world initially 
        """
        self.rows = rows
        self.cols = cols
        self.environment = environment
        self.animals = animals
        self.world_grid = None  ## make this in setup
        
        self.basic_setup(rows, cols, environment, animals)

    def basic_setup(self, rows, cols, environment, animals):
        """Sets up the world grid."""
        
        # make a variable that changes enviorment into something usefull
        env_order = ['Jungle', 'Forest', 'Meadow', 'Savannah', 'Desert']
        if environment in env_order:
            env_var = 5 - env_order.index(environment)
        else:
            env_var = 1  # default if invalid environment name

        # initlize world grid
        world_grid = np.zeros((rows, cols, 3))
        for x in range(rows):
            for y in range(cols):
                water = random.randint(3, 5) * (env_var /5)
                veg = random.randint(3, 5) * (env_var /5)
                meat = 0 
                world_grid[x, y] = [water, veg, meat]
        
        self.world_grid = world_grid


    ### TEST CODE FOR TESTING NOT FINAL DRAW FUNCTION ###
    def draw_board(self):
        """
        Draws the world board using matplotlib. Only draws Vegetation
        """
        layer = self.world_grid[:, :, 1] ## TAKES ONLY VALUES OF THE RELEVENT TYPE
        
        plt.figure(figsize=(6, 6))
        plt.imshow(layer, cmap='BrBG', vmin = 0, vmax = 5)
        plt.colorbar(label="Vegetation")
        plt.show()
```

```python
w = World(rows=20, cols=20, environment='Forest')
w.draw_board()
```
